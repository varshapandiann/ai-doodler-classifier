```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
classes = ["cat", "fish", "ice_cream", "swan", "tree"]
dataset_path = "../dataset"
```

```python
for cls in classes:
    file_path = os.path.join(dataset_path, f"{cls}.npy")
    
    if os.path.exists(file_path):
        data = np.load(file_path)
        print(f"Loaded '{cls}.npy' | Shape: {data.shape}")
        
        for i in range(5):
            sketch = data[i].reshape(28, 28)
            plt.imshow(sketch, cmap="gray")
            plt.title(f"{cls} Sketch #{i+1}")
            plt.axis("off")
            plt.show()
    else:
        print(f"File not found: {file_path}")
```
